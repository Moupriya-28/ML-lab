```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)


plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.title('KMeans Clustering on Iris (scikit-learn)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
# Compute SSE manually
import numpy as np # import numpy

def compute_sse(X, labels, centroids):
    sse = 0
    for i in range(len(centroids)): # get k from centroids length
        cluster_points = X[labels == i]
        sse += np.sum((cluster_points - centroids[i]) ** 2)
    return sse

centroids = kmeans.cluster_centers_ # define centroids
sse_manual = compute_sse(X, labels, centroids)
print("SSE using manual KMeans:", sse_manual)
```
