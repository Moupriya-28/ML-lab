```python
columns = ['BMA-101', 'BME-102', 'BCH-100', 'BCS-101', 'BCH-151', 'BCS-151','BCH-MC-002', 'BHM-100']
```

```python
def convert_grades(df, columns, grade_mapping):
    for col in columns:
        if col in df.columns:
            df[col] = df[col].map(grade_mapping)
    return df
```

```python
df = convert_grades(df.copy(), columns, grade_mapping)
```
